#### 1.2 回测函数导入

```python
def get_net(stock,start,end):
    dailyret = get_price(stock,start,end,'1d',['quote_rate'],is_panel =1)['quote_rate'].mean(axis =1)/100+1
    dailynet = dailyret.cumprod()
    return dailynet

def get_performance_analysis(T,year_day = 252):
    #输入净现值序列 
    #输出绩效指标
    
    #新高日期数
    max_T = 0
    #循环净值
    for s in range(2,len(T)):
        #节点划分
        l = T[:s]
        #判断当前节点为最大值
        if l[-1] > l[:-1].max():
        #新高日期+=1
            max_T += 1

    #净值新高占比数
    max_day_rate = max_T/(len(T)-1)
    max_day_rate = round(max_day_rate*100,2)
    
    #获取最终净值
    net_values = round(T[-1],4)
    
    #计算算数年华收益率
    year_ret_mean = T.pct_change().dropna().mean()*year_day
    year_ret_mean = round(year_ret_mean*100, 2)
    
    #计算几何年化收益率
    year_ret_sqrt = net_values ** (year_day / len(T) - 1)
    year_ret_sqrt = round(year_ret_sqrt**100 , 2)
    
    #计算年化波动率
    vol = T.pct_change().dropna().std()*np.sqrt(year_day)
    vol = round(vol*100,2)
    
    #计算夏普，无风险收益率3%
    sharpe = (year_ret_mean - 0.03)/ vol
    sharpe = round(sharpe,2)
    
    #计算最大回撤
    downlow = maxdrawdown(T)
    downlow = round(downlow*100,2)
    
    #输出
    return[net_values,year_ret_sqrt,downlow,sharpe,vol,max_day_rate]

#最大回测
# 再次定义函数：计算最大回撤
def maxdrawdown(arr):
    '''
    输入：净值序列
    输出：最大回撤
    '''
    # 最大回撤结束点
    i = np.argmax((np.maximum.accumulate(arr) - arr)/np.maximum.accumulate(arr))
    # 开始点
    j = np.argmax(arr[:i]) # start of period
    # 输出回撤值
    return (1-arr[i]/arr[j])
```

```python
#绘制布林带
MA =20                                                                    #可调
RANGE =2                                                                  #可调
BOMA = data_net.rolling(MA).mean()
UP = data_net.rolling(MA).mean()+data_net.rolling(MA).std()*RANGE
DN = data_net.rolling(MA).mean()+data_net.rolling(MA).std()*-RANGE
BO1 = pd.concat([data_net,BOMA,UP,DN,data_pct_change],axis =1)
BO1.columns = ['data_net_cum','BOMA','UP','DN','change']
BO1[['data_net_cum','BOMA','UP','DN']].plot(figsize=(16,10),title = '布林带')
```

```python
#截取信号部分
BO = BO1[MA:]
#计算信号
BO['diffup'] = BO['data_net_cum'] - BO['UP'] #大于零卖出-1
BO['diffdn'] =   BO['data_net_cum']- BO['DN'] #小于零买入1
x =0.00                                                   #可调优化的核心 信号敏感
BO['signal'] = np.where(BO['diffdn'] < -x, 1,0)
BO['signal'] = np.where(BO['diffup'] > x, -1,BO['signal'])
BO['signal'][0]=1
#0填补并填充
BO['signal'] = BO['signal'].replace(0,np.nan).fillna(method='ffill')
#把-1填充为0，清仓处理
BO['signal'] = BO['signal'].replace(-1,0)
#净值累计
BO['strategy_ret'] = (BO['signal']*BO['change']+1).cumprod()
BO['benchmark_ret'] = (BO['change']+1).cumprod()
BO[['benchmark_ret','strategy_ret']].plot(figsize=(16,10),title = '布林带策略回测')
```

```python
#截取信号部分
b = []
BO = BO1[MA:]
#计算信号
BO['diffup'] = BO['data_net_cum'] - BO['UP'] #大于零卖出-1
BO['diffdn'] =   BO['data_net_cum']- BO['DN'] #小于零买入1
for i in range(25,75):
    x =i/1000                                                   #可调优化的核心 信号敏感
    BO['signal'] = np.where(BO['diffdn'] < -x, 1,0)
    BO['signal'] = np.where(BO['diffup'] > x, -1,BO['signal'])
    BO['signal'][0]=1
    #0填补并填充
    BO['signal'] = BO['signal'].replace(0,np.nan).fillna(method='ffill')
    #把-1填充为0，清仓处理
    BO['signal'] = BO['signal'].replace(-1,0)
    #净值累计
    BO['strategy_ret'] = (BO['signal']*BO['change']+1).cumprod()
    b.append(BO.loc['2022-06-01','s'])
```

#### 2.4 计算单个股票的技术指标
+ talib快速计算技术指标，调用API：[talib](https://github.com/HuaRongSAO/talib-document)

计算指数移动平均值（EMA）<br>
+ 12日EMA的算式为 EMA（12）=前一日EMA（12）×11/13+今日收盘价×2/13
+ 26日EMA的算式为 EMA（26）=前一日EMA（26）×25/27+今日收盘价×2/27

计算离差值（DIF）<br>
+ DIF=今日EMA（12）－今日EMA（26）

DIF的9日EMA<br>
+ 根据离差值计算其9日的EMA，即离差平均值，是所求的MACD值。为了不与指标原名相混淆，此值又名DEA或DEM。
+ 今日DEA（MACD）=前一日DEA×8/10+今日DIF×2/10。
+ 计算出的DIF和DEA的数值均为正值或负值。<br>

故MACD指标是由两线一柱组合起来形成，快速线（白色线）为DIF，慢速线（黄色线）为DEA，柱状图为MACD。在各类投资中，有以下方法供投资者参考：

+ https://baike.baidu.com/item/EMA/12646151
+ https://zhuanlan.zhihu.com/p/41078267

```python
# talib技术指标库调用MACD
diff, dea, macd = talib.MACD(np.array(pricedata['close']), fastperiod=12, slowperiod=26, signalperiod=9)
# 创建dataFrame
dictdata = {'price':pricedata['close'],'DIFF': diff, 'DEA': dea, 'MACD':macd}
df = pd.DataFrame(dictdata,index = list(pricedata.index))
df = df.dropna()
df
```

```python
# 信号识别：MACD金叉买点，死叉卖点
def get_signal(datalist):
    if datalist[0] <0 and datalist[-1]>0:
        return 1
    elif datalist[0] >0 and datalist[-1]<0:
        return -1
    else:
        return 0

#运算
df['signal'] = df['MACD'].rolling(2).apply(lambda x:get_signal(x))
#去nan
df = df.dropna()
#展示
df.tail()
```

```python
def get_Profit_loss_ratio(x):
    l1 = [s if s > 0 else 0 for s in x]
    l2 = [s if s < 0 else 0 for s in x]
    return np.sum(l1)/abs(np.sum(l2))
buysignaldf['Profit loss ratio'] = buysignaldf['ret'].rolling(50,50).apply(lambda x:get_Profit_loss_ratio(x))
buysignaldf['Profit loss ratio'].plot(figsize = (12,3))
plt.title('MACD买入信号盈亏比走势（滚动50个信号',fontsize=15)
```
