# 3. 获取因子

#### 3.1 查询2022-01-05平安银行的MACD值

```python
q = query(
    factor.symbol,
    factor.date,
    factor.macd
    ).filter(
        factor.symbol == '000001.SZ',
        factor.date == '2022-01-05'                #try 2022-01-01
    )
get_factors(q)
```

#### 3.2  查询2022-01-05，MACD大于0，MACD从大到小排列的前10只股票

```python
q = query(
    factor.symbol,
    factor.date,
    factor.macd,
    ).filter(
    factor.macd > 0,
    factor.date == '2022-01-05'
    ).order_by(
        factor.macd.desc()
    ).limit(
        10
    )
get_factors(q)
```
